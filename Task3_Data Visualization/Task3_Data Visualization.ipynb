```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("countries.csv")
df.head()
```

```python
print(df.info())
```

Bar Chart

```python
top10 = df.sort_values(
    by="Population",
    ascending=False
).head(10)
plt.figure(figsize=(12,6))
plt.bar(
    top10["Country"],
    top10["Population"]
)
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```

Histogram

```python
plt.figure(figsize=(10,5))
plt.hist(
    df["Population"],
    bins=20
)
plt.title("Population Distribution")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

Box Plot (Seaborn)

```python
plt.figure(figsize=(10,4))
sns.boxplot(
    x=df["Population"]
)
plt.title("Population Outliers")
plt.show()
```

Pie Chart

```python
top5 = df.sort_values(
    by="Population",
    ascending=False
).head(5)
plt.figure(figsize=(8,8))

plt.pie(
    top5["Population"],
    labels=top5["Country"],
    autopct="%1.1f%%"
)
plt.title("Top 5 Population Share")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(10,5))
plt.scatter(
    range(len(df)),
    df["Population"]
)
plt.title("Population Scatter Plot")
plt.xlabel("Country Index")
plt.ylabel("Population")
plt.show()
```

# Data Story

This project analyzes country population data using visualizations to identify demographic patterns and trends.

The bar chart highlights the top 10 most populated countries. The histogram reveals that most countries have relatively small populations, while a few have exceptionally high populations. The box plot confirms the presence of population outliers, and the pie chart illustrates the population share of the five most populated countries.

These insights can help governments, researchers, and international organizations understand population distribution, allocate healthcare and education resources, and make informed policy decisions. Visualizations simplify complex data and enable faster, evidence-based decision-making.
